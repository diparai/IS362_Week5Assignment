2.What is the easternmost airport in the United States?

```python
east= pd.read_csv('airports.csv')
east_airport = east[(east['lat']> 44) & (east['lat'] <= 50) & (east['lon'] >= -75) & (east['lon'] <= -65)]
east_airport.head()
east_airport.sort_values('lon', ascending=False).head(5)
```

 **Eastport Municipal Airport** is the earthernmost airport in the United States. I double checked the longitude and latitude in multiple websites and it is correct as mentioned above.

3. On February 12th, 2013, which New York area airport had the windiest weather?


```python
windy = pd.read_csv('weather.csv')
# sorting data based on requirements
windy_weather = windy[(windy["month"]==2) & (windy["day"]==12) & (windy["year"]==2013)]
windy_weather.head()
windy_weather.sort_values("wind_speed", ascending=False).head(6)
```
